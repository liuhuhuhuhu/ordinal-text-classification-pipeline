```python
# Standard libraries
import re
import html
from pathlib import Path

# Data processing
import pandas as pd
import numpy as np

# Input and output paths
RAW_PATH = "posts_raw.csv"
OUT_DIR = Path("outputs")
OUT_DIR.mkdir(exist_ok=True)
```

```python
## Part 1 — Data engineering
```

```python
# Load raw dataset without modification.
df = pd.read_csv(RAW_PATH)

print("Raw dataset shape:", df.shape)
df.head()
```

```python
# Initial data audit

# Before cleaning, inspect:
# 1. Missingness patterns
# 2. Label formats
# 3. Duplicate observations
# 4. Potential inconsistencies
print(df.info())
print(df.isna().sum())

print("\nRaw label distribution:")
print(df["label"].value_counts(dropna=False))

print("\nDuplicate post_id:", df["post_id"].duplicated().sum())
print("Duplicate text:", df["text"].duplicated().sum())
```

```python
# Step 2: Standardize sentiment labels to the canonical 0-4 scale

# The raw labels are intentionally heterogeneous and may appear as integers, floats, abbreviations, textual labels, or missing values. I convert all valid labels to the common ordinal scale:

# 0 = very negative
# 1 = negative
# 2 = neutral
# 3 = positive
# 4 = very positive

# Unrecognized labels are treated as missing values and removed
# later in the cleaning pipeline.


def canonicalize_label(x):
    """
    Convert heterogeneous sentiment labels to a standardized
    ordinal scale ranging from 0 to 4.

    Parameters
    ----------
    x : object
        Raw label value from the original dataset.

    Returns
    -------
    int or np.nan
        Integer sentiment label in {0,1,2,3,4} or missing value
        if the label cannot be interpreted.
    """

    # Explicit missing values
    if pd.isna(x):
        return np.nan

    # Normalize capitalization and surrounding whitespace
    s = str(x).strip().lower()

    # Common placeholders for missing labels
    if s in {"", "?", "nan", "none", "missing"}:
        return np.nan


    # Handle numeric labels such as:
    # 0, 1, 2, 3, 4, 3.0, 4.0
    try:
        value = float(s)

        if value.is_integer() and int(value) in {0,1,2,3,4}:
            return int(value)

    except ValueError:
        pass

    # Handle textual sentiment labels and abbreviations

    label_map = {
        "vneg": 0,
        "very negative": 0,

        "neg": 1,
        "negative": 1,

        "neu": 2,
        "neutral": 2,

        "pos": 3,
        "positive": 3,

        "vpos": 4,
        "very positive": 4
    }

    return label_map.get(s, np.nan)


# Apply label harmonization
df["label_clean"] = df["label"].apply(canonicalize_label)

# Inspect the resulting label distribution
print(df["label_clean"].value_counts(dropna=False).sort_index())
```

```python
# Step 3: Light text normalization

#   - decode HTML entities;
#   - normalize escaped characters;
#   - replace URLs and user mentions with placeholders;
#   - standardize whitespace.

# I intentionally avoid aggressive preprocessing such as:
#   - stopword removal
#   - stemming
#   - lemmatization
#
# because negations and function words often carry sentiment information in short social media texts.


def clean_text(text):
    """
    Perform light normalization while preserving linguistic
    signals relevant for sentiment classification.
    """

    if pd.isna(text):
        return np.nan

    s = str(text)

    s = html.unescape(s)
    s = s.replace("\\/", "/")

    # Preserve structural information using placeholders
    s = re.sub(r"https?://\S+|www\.\S+", " URL ", s)
    s = re.sub(r"@\w+", " USER ", s)

    # Normalize whitespace introduced during scraping
    s = re.sub(r"\s+", " ", s).strip()

    return s


df["text_clean"] = df["text"].apply(clean_text)
```

```python
# Step 4: Maintain an audit trail for cleaning decisions

# To ensure reproducibility and transparency, I document:

#   - remaining observations after each step;
#   - missing values;
#   - the number of unique texts.


audit = []

def log_step(name, data):

    audit.append({
        "step": name,
        "rows": len(data),

        "unique_texts":
            data["text_clean"].nunique()
            if "text_clean" in data.columns else np.nan,

        "missing_labels":
            data["label_clean"].isna().sum()
            if "label_clean" in data.columns else np.nan,

        "missing_texts":
            data["text_clean"].isna().sum()
            if "text_clean" in data.columns else np.nan
    })


log_step("raw_loaded", df)

clean = df.copy()

# Remove missing and unrecognized labels
clean = clean.dropna(subset=["label_clean"])
log_step("drop_missing_or_unrecognized_labels", clean)

# Remove missing texts
clean = clean.dropna(subset=["text_clean"])
clean = clean[clean["text_clean"].str.len() > 0]
log_step("drop_missing_or_empty_text", clean)
```

```python
# Step 5: Handle duplicate texts and annotation conflicts


# Particular attention is given to duplicate texts with conflicting labels because they imply disagreement in the supervision signal.
#
# I adopt a relatively conservative strategy:
#
#   - if duplicate texts share the same label, keep one copy;
#   - if duplicate texts have conflicting labels, remove the
#     entire duplicate group.
#
# This prioritizes label quality over sample size and avoids
# training on contradictory supervision signals.
# ==========================================================

label_nunique_by_text = (
    clean.groupby("text_clean")["label_clean"]
    .nunique()
)

conflicting_texts = (
    label_nunique_by_text[
        label_nunique_by_text > 1
    ].index
)

print(
    "Number of conflicting duplicate groups:",
    len(conflicting_texts)
)

print(
    "Rows involved in conflicts:",
    clean["text_clean"].isin(conflicting_texts).sum()
)

# Save representative examples for reporting
conflict_examples = (
    clean[
        clean["text_clean"].isin(conflicting_texts)
    ]
    .sort_values("text_clean")
)

# Remove conflicting duplicates
clean = clean[
    ~clean["text_clean"].isin(conflicting_texts)
].copy()

log_step("drop_conflicting_duplicate_texts", clean)

# Keep one representative copy for consistent duplicates
clean = (
    clean
    .sort_values(["text_clean", "post_id"])
    .drop_duplicates(
        subset=["text_clean"],
        keep="first"
    )
)

log_step("deduplicate_consistent_texts", clean)
```

```python
# Step 6: Export modeling-ready dataset

# The final dataset contains:
#
#   - harmonized ordinal labels;
#   - normalized texts;
#   - one observation per unique text instance.

# Auxiliary outputs are exported to support reproducibility and reporting.


final = clean[
    [
        "post_id",
        "text_clean",
        "label_clean",
        "source",
        "collected_at"
    ]
].copy()

final = final.rename(
    columns={
        "text_clean": "text",
        "label_clean": "label"
    }
)

final["label"] = final["label"].astype(int)

# Save cleaned dataset
final.to_csv(
    OUT_DIR / "posts_clean.csv",
    index=False
)

# Save cleaning audit table
audit_df = pd.DataFrame(audit)

audit_df.to_csv(
    OUT_DIR / "cleaning_audit.csv",
    index=False
)

# Save class distribution statistics
class_dist = (
    final["label"]
    .value_counts()
    .sort_index()
    .reset_index()
)

class_dist.columns = [
    "label",
    "count"
]

class_dist["share"] = (
    class_dist["count"]
    / class_dist["count"].sum()
)

class_dist.to_csv(
    OUT_DIR / "class_distribution.csv",
    index=False
)

print(audit_df)
print(class_dist)
```

```python
# ==========================================================
# Part 2: Baseline model
# ==========================================================
#
# Objective:
# Train a classical NLP baseline using sparse text features
# and a linear classifier.
#
# This model serves as the reference point for the later
# ordinal-aware continuous scoring model.
#
# Baseline specification:
#   - TF-IDF features
#   - Logistic Regression classifier
#   - Stratified train/validation split
#   - Fixed random seed for reproducibility
# ==========================================================

import pandas as pd
import numpy as np

from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    mean_absolute_error,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python

# Step 1: Load cleaned modeling data
# I use the cleaned dataset produced in Part 1.


DATA_PATH = Path("outputs/posts_clean.csv")
OUT_DIR = Path("outputs")
OUT_DIR.mkdir(exist_ok=True)

df = pd.read_csv(DATA_PATH)

print("Cleaned dataset shape:", df.shape)
print(df.head())

print("\nClass distribution:")
print(df["label"].value_counts().sort_index())
```

```python
# Step 2: Create stratified validation split

SEED = 42

X = df["text"].astype(str)
y = df["label"].astype(int)

X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=SEED
)

print("Training size:", len(X_train))
print("Validation size:", len(X_val))

print("\nTraining label distribution:")
print(y_train.value_counts(normalize=True).sort_index())

print("\nValidation label distribution:")
print(y_val.value_counts(normalize=True).sort_index())
```

```python
# Step 3: Build TF-IDF feature representation

# min_df removes extremely rare terms that are unlikely to generalize well.


tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),
    min_df=3,
    max_df=0.95,
    max_features=30000,
    sublinear_tf=True
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_val_tfidf = tfidf.transform(X_val)

print("TF-IDF training matrix shape:", X_train_tfidf.shape)
print("TF-IDF validation matrix shape:", X_val_tfidf.shape)
```

```python
# Step 4: Train linear baseline classifier using Logistic Regression

baseline_clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    solver="liblinear",
    random_state=SEED
)

baseline_clf.fit(X_train_tfidf, y_train)

y_val_pred = baseline_clf.predict(X_val_tfidf)

print("Baseline validation predictions generated.")
```

```python
# Step 5: Define ordinal-aware evaluation metrics

# Within-one accuracy:
#   share of predictions where |pred - true| <= 1
#
# Signed bias:
#   mean(pred - true)
# A negative signed bias means the model tends to underpredict sentiment; a positive signed bias means it tends to overpredict.

def evaluate_ordinal_classification(y_true, y_pred, labels=[0, 1, 2, 3, 4]):
    """
    Evaluate model predictions using the lab's standard
    ordinal classification metric suite.

    Parameters
    ----------
    y_true : array-like
        True ordinal labels.
    y_pred : array-like
        Predicted ordinal labels.
    labels : list
        Ordered class labels.

    Returns
    -------
    metrics : dict
        Dictionary containing scalar metrics, per-class F1,
        and the confusion matrix.
    """

    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    exact_accuracy = accuracy_score(y_true, y_pred)

    within_one_accuracy = np.mean(
        np.abs(y_pred - y_true) <= 1
    )

    mae = mean_absolute_error(y_true, y_pred)

    signed_bias = np.mean(y_pred - y_true)

    macro_f1 = f1_score(
        y_true,
        y_pred,
        labels=labels,
        average="macro",
        zero_division=0
    )

    per_class_f1 = f1_score(
        y_true,
        y_pred,
        labels=labels,
        average=None,
        zero_division=0
    )

    cm = confusion_matrix(
        y_true,
        y_pred,
        labels=labels
    )

    metrics = {
        "exact_accuracy": exact_accuracy,
        "within_one_accuracy": within_one_accuracy,
        "mae": mae,
        "signed_bias": signed_bias,
        "macro_f1": macro_f1,
        "per_class_f1": per_class_f1,
        "confusion_matrix": cm
    }

    return metrics
```

```python
# Step 6: Evaluate baseline on validation set


baseline_metrics = evaluate_ordinal_classification(
    y_true=y_val,
    y_pred=y_val_pred
)

print("Exact accuracy:", round(baseline_metrics["exact_accuracy"], 4))
print("Within-one accuracy:", round(baseline_metrics["within_one_accuracy"], 4))
print("MAE:", round(baseline_metrics["mae"], 4))
print("Signed bias:", round(baseline_metrics["signed_bias"], 4))
print("Macro-F1:", round(baseline_metrics["macro_f1"], 4))

print("\nPer-class F1:")
for label, score in zip([0, 1, 2, 3, 4], baseline_metrics["per_class_f1"]):
    print(f"Label {label}: {score:.4f}")

print("\nConfusion matrix:")
print(baseline_metrics["confusion_matrix"])

print("\nClassification report:")
print(
    classification_report(
        y_val,
        y_val_pred,
        labels=[0, 1, 2, 3, 4],
        zero_division=0
    )
)
```

```python

# Step 7: Save baseline metrics for reporting


baseline_scalar_metrics = pd.DataFrame([
    {
        "model": "tfidf_logistic_regression",
        "exact_accuracy": baseline_metrics["exact_accuracy"],
        "within_one_accuracy": baseline_metrics["within_one_accuracy"],
        "mae": baseline_metrics["mae"],
        "signed_bias": baseline_metrics["signed_bias"],
        "macro_f1": baseline_metrics["macro_f1"]
    }
])

baseline_scalar_metrics.to_csv(
    OUT_DIR / "baseline_validation_metrics.csv",
    index=False
)

baseline_per_class_f1 = pd.DataFrame({
    "label": [0, 1, 2, 3, 4],
    "f1": baseline_metrics["per_class_f1"]
})

baseline_per_class_f1.to_csv(
    OUT_DIR / "baseline_per_class_f1.csv",
    index=False
)

baseline_confusion_matrix = pd.DataFrame(
    baseline_metrics["confusion_matrix"],
    index=[f"true_{i}" for i in range(5)],
    columns=[f"pred_{i}" for i in range(5)]
)

baseline_confusion_matrix.to_csv(
    OUT_DIR / "baseline_confusion_matrix.csv"
)

print("Saved baseline outputs to outputs/.")
```

```python

# Part 3: Continuous scoring + threshold calibration

# I use class probabilities from TF-IDF + Logistic Regression
# and convert them into a continuous score using the expected value:

# score = sum_k P(y=k | x) * k


import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    mean_absolute_error,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python

# Step 1: Load cleaned data


SEED = 42
OUT_DIR = Path("outputs")
OUT_DIR.mkdir(exist_ok=True)

df = pd.read_csv("outputs/posts_clean.csv")

X = df["text"].astype(str)
y = df["label"].astype(int)

print("Dataset shape:", df.shape)
print(y.value_counts().sort_index())
```

```python

# Step 2: Recreate the same validation split as Part 2

X_train_full, X_val, y_train_full, y_val = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=SEED
)

print("Full training size:", len(X_train_full))
print("Validation size:", len(X_val))
```

```python

# Step 3: Create held-out calibration split

X_train_model, X_cal, y_train_model, y_cal = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.20,
    stratify=y_train_full,
    random_state=SEED
)

print("Model training size:", len(X_train_model))
print("Calibration size:", len(X_cal))
print("Validation size:", len(X_val))
```

```python

# Step 4: Define evaluation function

# This is the same metric suite used in Part 2.
# Within-one accuracy and signed bias are implemented directly because they are ordinal metrics not directly provided by sklearn.


def evaluate_ordinal_classification(y_true, y_pred, labels=[0, 1, 2, 3, 4]):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    return {
        "exact_accuracy": accuracy_score(y_true, y_pred),
        "within_one_accuracy": np.mean(np.abs(y_pred - y_true) <= 1),
        "mae": mean_absolute_error(y_true, y_pred),
        "signed_bias": np.mean(y_pred - y_true),
        "macro_f1": f1_score(
            y_true,
            y_pred,
            labels=labels,
            average="macro",
            zero_division=0
        ),
        "per_class_f1": f1_score(
            y_true,
            y_pred,
            labels=labels,
            average=None,
            zero_division=0
        ),
        "confusion_matrix": confusion_matrix(
            y_true,
            y_pred,
            labels=labels
        )
    }
```

```python

# Step 5: Fit Part 2 baseline on the full training set

# This recreates the Part 2 baseline for comparison on the same validation split.


baseline_tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),
    min_df=3,
    max_df=0.95,
    max_features=30000,
    sublinear_tf=True
)

X_train_full_tfidf = baseline_tfidf.fit_transform(X_train_full)
X_val_baseline_tfidf = baseline_tfidf.transform(X_val)

baseline_clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    solver="liblinear",
    random_state=SEED
)

baseline_clf.fit(X_train_full_tfidf, y_train_full)

y_val_baseline_pred = baseline_clf.predict(X_val_baseline_tfidf)

baseline_metrics = evaluate_ordinal_classification(
    y_val,
    y_val_baseline_pred
)

print("Baseline exact accuracy:", baseline_metrics["exact_accuracy"])
```

```python

# Step 6: Fit continuous-score model

# The continuous model is also based on TF-IDF + Logistic Regression, but instead of taking the argmax class, we use predicted class probabilities to compute an expected ordinal score.
#
# score_i = sum_k P(y_i = k | x_i) * k


ordinal_tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),
    min_df=3,
    max_df=0.95,
    max_features=30000,
    sublinear_tf=True
)

X_train_model_tfidf = ordinal_tfidf.fit_transform(X_train_model)
X_cal_tfidf = ordinal_tfidf.transform(X_cal)
X_val_ordinal_tfidf = ordinal_tfidf.transform(X_val)

ordinal_clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    solver="liblinear",
    random_state=SEED
)

ordinal_clf.fit(X_train_model_tfidf, y_train_model)

class_values = np.array([0, 1, 2, 3, 4])

cal_proba = ordinal_clf.predict_proba(X_cal_tfidf)
val_proba = ordinal_clf.predict_proba(X_val_ordinal_tfidf)

cal_scores = cal_proba @ class_values
val_scores = val_proba @ class_values

print("Example continuous scores:")
print(cal_scores[:10])
```

```python

# Step 7: Naive rounding benchmark

def round_scores_to_labels(scores):
    return np.clip(
        np.rint(scores),
        0,
        4
    ).astype(int)


y_val_round_pred = round_scores_to_labels(val_scores)

rounding_metrics = evaluate_ordinal_classification(
    y_val,
    y_val_round_pred
)

print("Naive rounding exact accuracy:", rounding_metrics["exact_accuracy"])
```

```python

# Step 8: Threshold search using dynamic programming

# search for four cut-point thresholds that maximize exact accuracy on the held-out calibration set.
# Instead of using arbitrary thresholds such as 0.5, 1.5, 2.5,

# The algorithm sorts examples by continuous score and finds the best partition of the ordered calibration examples into five contiguous bins corresponding to labels 0,1,2,3,4.


def fit_thresholds_dp(scores, y_true, labels=[0, 1, 2, 3, 4]):
    scores = np.asarray(scores)
    y_true = np.asarray(y_true)

    order = np.argsort(scores)
    scores_sorted = scores[order]
    y_sorted = y_true[order]

    n = len(scores_sorted)
    k = len(labels)

    # prefix_counts[c, i] = number of observations with label c
    # among the first i sorted observations
    prefix_counts = np.zeros((k, n + 1), dtype=int)

    for i in range(n):
        prefix_counts[:, i + 1] = prefix_counts[:, i]
        label_index = labels.index(y_sorted[i])
        prefix_counts[label_index, i + 1] += 1

    def segment_correct(label_idx, start, end):
        """
        Number of correct predictions if observations from
        start to end are assigned to labels[label_idx].
        """
        return prefix_counts[label_idx, end] - prefix_counts[label_idx, start]

    # dp[j, i] = best correct count using first i observations
    # assigned into first j classes
    dp = np.full((k + 1, n + 1), -np.inf)
    back = np.zeros((k + 1, n + 1), dtype=int)

    dp[0, 0] = 0

    for j in range(1, k + 1):
        label_idx = j - 1

        for i in range(j, n + 1):
            best_score = -np.inf
            best_m = None

            # m is the end of the previous class segment
            for m in range(j - 1, i):
                candidate = (
                    dp[j - 1, m]
                    + segment_correct(label_idx, m, i)
                )

                if candidate > best_score:
                    best_score = candidate
                    best_m = m

            dp[j, i] = best_score
            back[j, i] = best_m

    # Recover split points
    split_points = []
    i = n

    for j in range(k, 0, -1):
        m = back[j, i]
        split_points.append(m)
        i = m

    split_points = list(reversed(split_points))[1:]

    thresholds = []

    for split in split_points:
        if split <= 0:
            threshold = scores_sorted[0] - 1e-6
        elif split >= n:
            threshold = scores_sorted[-1] + 1e-6
        else:
            threshold = (
                scores_sorted[split - 1]
                + scores_sorted[split]
            ) / 2

        thresholds.append(threshold)

    return np.array(thresholds)
```

```python

# Step 9: Apply calibrated thresholds


def apply_thresholds(scores, thresholds):
    """
    Convert continuous scores into ordinal class labels using
    fitted cut-point thresholds.
    """
    return np.digitize(scores, thresholds).astype(int)


thresholds = fit_thresholds_dp(
    scores=cal_scores,
    y_true=y_cal.to_numpy(),
    labels=[0, 1, 2, 3, 4]
)

print("Fitted thresholds:")
print(thresholds)

y_cal_calibrated_pred = apply_thresholds(
    cal_scores,
    thresholds
)

y_val_calibrated_pred = apply_thresholds(
    val_scores,
    thresholds
)

calibrated_metrics = evaluate_ordinal_classification(
    y_val,
    y_val_calibrated_pred
)

print("Calibrated exact accuracy:", calibrated_metrics["exact_accuracy"])
```

```python

# Step 10A: Alternative continuous-score model
# TF-IDF + Ridge Regression

# The expected-value method above uses class probabilities from a classifier. As an alternative, I directly train a regression model to predict the ordinal label as a continuous outcome.


from sklearn.linear_model import Ridge

ridge_tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),
    min_df=3,
    max_df=0.95,
    max_features=30000,
    sublinear_tf=True
)

X_train_model_ridge = ridge_tfidf.fit_transform(X_train_model)
X_cal_ridge = ridge_tfidf.transform(X_cal)
X_val_ridge = ridge_tfidf.transform(X_val)

ridge_reg = Ridge(
    alpha=1.0,
    random_state=SEED
)

ridge_reg.fit(
    X_train_model_ridge,
    y_train_model
)

cal_scores_ridge = ridge_reg.predict(X_cal_ridge)
val_scores_ridge = ridge_reg.predict(X_val_ridge)

# Keep scores within the valid ordinal range for interpretability
cal_scores_ridge = np.clip(cal_scores_ridge, 0, 4)
val_scores_ridge = np.clip(val_scores_ridge, 0, 4)

print("Example Ridge continuous scores:")
print(val_scores_ridge[:10])

print("\nRidge score summary:")
print(pd.Series(val_scores_ridge).describe())
```

```python

# Step 10B: Naive rounding for Ridge scores


y_val_ridge_round_pred = round_scores_to_labels(
    val_scores_ridge
)

ridge_rounding_metrics = evaluate_ordinal_classification(
    y_val,
    y_val_ridge_round_pred
)

print("Ridge naive rounding exact accuracy:",
      ridge_rounding_metrics["exact_accuracy"])

print("Ridge naive rounding within-one accuracy:",
      ridge_rounding_metrics["within_one_accuracy"])

print("Ridge naive rounding MAE:",
      ridge_rounding_metrics["mae"])
```

```python

# Step 12: Cumulative Binary Ordinal Classifiers

# This method decomposes the ordinal 5-class problem into
# four cumulative binary tasks:

#   P(y > 0), P(y > 1), P(y > 2), P(y > 3)

# The continuous score is:
#
#   score = P(y>0) + P(y>1) + P(y>2) + P(y>3)
#
# I additionally enforce monotonicity:
#
#   P(y>0) >= P(y>1) >= P(y>2) >= P(y>3)


from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.base import clone


binary_tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.95,
    max_features=50000,
    sublinear_tf=True
)

X_train_binary = binary_tfidf.fit_transform(X_train_model)
X_cal_binary = binary_tfidf.transform(X_cal)
X_val_binary = binary_tfidf.transform(X_val)

threshold_levels = [0, 1, 2, 3]
binary_models = {}

# Train four cumulative classifiers


for t in threshold_levels:
    y_train_binary = (y_train_model > t).astype(int)

    base_clf = LogisticRegression(
        max_iter=3000,
        class_weight="balanced",
        solver="liblinear",
        C=2.0,
        random_state=SEED
    )

    base_clf.fit(X_train_binary, y_train_binary)
    binary_models[t] = base_clf

print("Optimized cumulative binary classifiers trained.")
```

```python

# Generate monotonic cumulative probabilities

def cumulative_probs(models, X_matrix, thresholds=[0, 1, 2, 3]):
    """
    Return an n x 4 matrix of cumulative probabilities:
    P(y>0), P(y>1), P(y>2), P(y>3).
    """

    probs = []

    for t in thresholds:
        p = models[t].predict_proba(X_matrix)[:, 1]
        probs.append(p)

    probs = np.vstack(probs).T

    return probs


def enforce_monotonicity(probs):
    """
    Enforce ordinal consistency:
    P(y>0) >= P(y>1) >= P(y>2) >= P(y>3).

    We use a simple cumulative minimum from left to right.
    """

    probs_mono = probs.copy()

    for j in range(1, probs_mono.shape[1]):
        probs_mono[:, j] = np.minimum(
            probs_mono[:, j - 1],
            probs_mono[:, j]
        )

    return probs_mono


cal_probs_raw = cumulative_probs(binary_models, X_cal_binary)
val_probs_raw = cumulative_probs(binary_models, X_val_binary)

cal_probs_mono = enforce_monotonicity(cal_probs_raw)
val_probs_mono = enforce_monotonicity(val_probs_raw)

cal_scores_binary_opt = cal_probs_mono.sum(axis=1)
val_scores_binary_opt = val_probs_mono.sum(axis=1)

print("Example optimized cumulative scores:")
print(val_scores_binary_opt[:10])

print("\nOptimized score summary:")
print(pd.Series(val_scores_binary_opt).describe())
```

```python

# Option A: Direct ordinal prediction by counting thresholds


# A direct class prediction can be obtained by counting how
# many cumulative probabilities exceed 0.5.

y_val_binary_direct_pred = (val_probs_mono >= 0.5).sum(axis=1)

binary_direct_metrics = evaluate_ordinal_classification(
    y_val,
    y_val_binary_direct_pred
)

print("Cumulative binary direct exact accuracy:",
      binary_direct_metrics["exact_accuracy"])

print("Cumulative binary direct within-one accuracy:",
      binary_direct_metrics["within_one_accuracy"])

print("Cumulative binary direct MAE:",
      binary_direct_metrics["mae"])

print("Cumulative binary direct signed bias:",
      binary_direct_metrics["signed_bias"])

print("Cumulative binary direct macro-F1:",
      binary_direct_metrics["macro_f1"])
```

```python

# Optimized Part 3:
# all-mpnet-base-v2 + Ridge alpha search + threshold calibration

# This is an embedding-based continuous scoring model.

# Steps:
#   1. Encode texts using all-mpnet-base-v2 sentence embeddings
#   2. Train Ridge regression models with different alpha values
#   3. Select alpha using the calibration split only
#   4. Fit thresholds on the calibration split
#   5. Evaluate final model on the validation split


!pip install -q sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.linear_model import Ridge
import pandas as pd
import numpy as np
```

```python

# Step 1: Load stronger sentence embedding model

# all-mpnet-base-v2 is slower than all-MiniLM-L6-v2 but often provides stronger semantic representations.


mpnet_model = SentenceTransformer(
    "sentence-transformers/all-mpnet-base-v2"
)

print("MPNet embedding model loaded.")
```

```python

# Step 2: Generate MPNet embeddings


X_train_mpnet = mpnet_model.encode(
    X_train_model.tolist(),
    show_progress_bar=True,
    convert_to_numpy=True,
    batch_size=32
)

X_cal_mpnet = mpnet_model.encode(
    X_cal.tolist(),
    show_progress_bar=True,
    convert_to_numpy=True,
    batch_size=32
)

X_val_mpnet = mpnet_model.encode(
    X_val.tolist(),
    show_progress_bar=True,
    convert_to_numpy=True,
    batch_size=32
)

print("Training embedding shape:", X_train_mpnet.shape)
print("Calibration embedding shape:", X_cal_mpnet.shape)
print("Validation embedding shape:", X_val_mpnet.shape)
```

```python

# Step 3: Ridge alpha search on calibration set

# Selection criterion:
#   1. maximize calibrated exact accuracy on calibration set;
#   2. break ties using lower calibration MAE.


alpha_grid = [0.01, 0.03, 0.1, 0.3, 1.0, 3.0, 10.0, 30.0, 100.0]

alpha_results = []

for alpha in alpha_grid:
    ridge = Ridge(
        alpha=alpha,
        random_state=SEED
    )

    ridge.fit(
        X_train_mpnet,
        y_train_model
    )

    cal_scores = ridge.predict(X_cal_mpnet)

    # Keep scores interpretable on the 0-4 ordinal scale
    cal_scores = np.clip(cal_scores, 0, 4)

    # Fit thresholds on calibration scores only
    thresholds_alpha = fit_thresholds_dp(
        scores=cal_scores,
        y_true=y_cal.to_numpy(),
        labels=[0, 1, 2, 3, 4]
    )

    cal_pred = apply_thresholds(
        cal_scores,
        thresholds_alpha
    )

    cal_metrics = evaluate_ordinal_classification(
        y_cal,
        cal_pred
    )

    alpha_results.append({
        "alpha": alpha,
        "cal_exact_accuracy": cal_metrics["exact_accuracy"],
        "cal_within_one_accuracy": cal_metrics["within_one_accuracy"],
        "cal_mae": cal_metrics["mae"],
        "cal_signed_bias": cal_metrics["signed_bias"],
        "cal_macro_f1": cal_metrics["macro_f1"],
        "thresholds": thresholds_alpha
    })

alpha_results_df = pd.DataFrame(alpha_results)

display(
    alpha_results_df.drop(columns=["thresholds"])
    .sort_values(
        ["cal_exact_accuracy", "cal_mae"],
        ascending=[False, True]
    )
)
```

```python

# Step 4: Select best alpha from calibration set


best_row = (
    alpha_results_df
    .sort_values(
        ["cal_exact_accuracy", "cal_mae"],
        ascending=[False, True]
    )
    .iloc[0]
)

best_alpha = best_row["alpha"]
best_mpnet_thresholds = best_row["thresholds"]

print("Best alpha selected on calibration set:", best_alpha)
print("Best calibration thresholds:")
print(best_mpnet_thresholds)
```

```python

# Step 5: Train final MPNet + Ridge model with selected alpha


mpnet_ridge = Ridge(
    alpha=best_alpha,
    random_state=SEED
)

mpnet_ridge.fit(
    X_train_mpnet,
    y_train_model
)

cal_scores_mpnet = mpnet_ridge.predict(X_cal_mpnet)
val_scores_mpnet = mpnet_ridge.predict(X_val_mpnet)

cal_scores_mpnet = np.clip(cal_scores_mpnet, 0, 4)
val_scores_mpnet = np.clip(val_scores_mpnet, 0, 4)

print("Example MPNet continuous scores:")
print(val_scores_mpnet[:10])

print("\nMPNet score summary:")
print(pd.Series(val_scores_mpnet).describe())
```
